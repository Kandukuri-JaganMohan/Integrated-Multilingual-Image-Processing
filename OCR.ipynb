```python
def initialize_models():
  det_model = db_resnet50(pretrained=True)
  reco_model = crnn_vgg16_bn(pretrained=True, pretrained_backbone=True,vocab=VOCABS['hindi'])
  reco_dict = torch.load("/content/drive/MyDrive/Project/Models/crnn_vgg16_bn_handwritten_hindi.pt",map_location="cpu")
  reco_model.load_state_dict(reco_dict)

  det_predictor = DetectionPredictor(
    PreProcessor(
        (1024, 1024),
        batch_size=1,
        mean=(0.798, 0.785, 0.772),
        std=(0.264, 0.2749, 0.287)
    ),
    det_model
  )

  reco_predictor = RecognitionPredictor(
    PreProcessor(
        (32, 128),
        preserve_aspect_ratio=True,
        batch_size=32,
        mean=(0.694, 0.695, 0.693),
        std=(0.299, 0.296, 0.301)
    ),
    reco_model
  )
  predictor = OCRPredictor(det_predictor, reco_predictor)
  return predictor
```

```python
def get_txt(json_op,img):
  result=''
  for page in json_op['pages']:
    for block in page['blocks']:
      for lines in block['lines']:
        for word in lines['words']:
          result+=word['value']+' '
          result+='\n'
  return results(img)
```

```python
def predict(file_path):
  doc = DocumentFile.from_images(file_path)
  img = Image.open(file_path)
  model=initialize_models()
  result = model(doc)
  output = get_txt(result.export(),img)
  return output
```

```python
def initialize_models(lang):
  models_dict = {"telugu":"Models/crnn_vgg16_bn_telugu.pt","hindi":"Models/crnn_vgg16_bn_hindi.pt","kannada":"Models/crnn_vgg16_bn_kannada.pt","gujarati":"Models/crnn_vgg16_bn_gujarati.pt"}
  det_model = db_resnet50(pretrained=True)
  reco_model = crnn_vgg16_bn(pretrained=False, pretrained_backbone=False,vocab=VOCABS[lang])
  reco_dict = torch.load(models_dict[lang],map_location="cpu")
  reco_model.load_state_dict(reco_dict)

  det_predictor = DetectionPredictor(
    PreProcessor(
        (1024, 1024),
        batch_size=1,
        mean=(0.798, 0.785, 0.772),
        std=(0.264, 0.2749, 0.287)
    ),
    det_model
  )

  reco_predictor = RecognitionPredictor(
    PreProcessor(
        (32, 128),
        preserve_aspect_ratio=True,
        batch_size=32,
        mean=(0.694, 0.695, 0.693),
        std=(0.299, 0.296, 0.301)
    ),
    reco_model
  )
  predictor = OCRPredictor(det_predictor, reco_predictor)
  return predictor
```
