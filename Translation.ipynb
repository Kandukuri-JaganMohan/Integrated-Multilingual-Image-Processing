```python
from deep_translator import GoogleTranslator
```

```python
translator = GoogleTranslator(source='auto')
```

```python
lang_codes = translator.get_supported_languages(as_dict=True)
lang_codes
```

```python
langs = translator.get_supported_languages(as_dict=False)
inp_langs = "\n"
for i in range(len(langs)):
  inp_langs+=str(i+1)+"."
  inp_langs+=langs[i]
  inp_langs+="\n"
inp_langs
```

```python
text = input("Enter text: ")
language = input("Enter target language: "+inp_langs)
target_lang = lang_codes[language]
target_lang
```

```python
# text = 'I want to translate this text'
# target_lang = 'de'
translated = GoogleTranslator(source='auto', target=target_lang).translate(text)
translated
```
