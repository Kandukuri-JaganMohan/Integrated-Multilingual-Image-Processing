## GTTS for Foreign languages

```python
from gtts import gTTS
langs = {"English":{"en":{"Australia":"com.au","United Kingdom":"co.uk","United States":"us","Canada":"ca","India":"co.in","Ireland":"ie","South Africa":"co.za"}},
         "French":{"fr":{"Canada":"ca","France":"fr"}},
         "Portuguese":{"pt":{"Brazil":"com.br","Portugal":"pt"}},
         "Spanish":{"es":{"Mexico":"com.mx","Spain":"es","United States":"us"}}
         }
lang_str = ""
for i in list(langs.keys()):
  lang_str+=str(i)+"\n"
# sample = input("Enter text to set voice alert: ")
try:
  c=0
  language = input(lang_str+"Select language: ")
  lang = list(langs[language].keys())[0]
  print("=================================================")
  # print(lang)
  c=1
  accent_str = ""
  for i in list(langs[language][lang].keys()):
    accent_str+=str(i)+"\n"
  # print(accent_str)
  accent = input(accent_str+"Select accent: ")
  tld = langs[language][lang][accent]
  print("=================================================")
  text = input("Enter text to set voice alert: ")
except:
  if(c==0):
    print("Select one of the given languages only")
  if(c==1):
    print("Select one of the given accents only")
myobj = gTTS(text=text,lang=lang,tld=tld,slow=False)
myobj.save("es-us.mp3")
```

```python
from IPython.display import HTML

def open_in_new_tab(url):
  """Opens the given URL in a new tab."""
  html = """
  <script>
  window.open("{url}");
  </script>
  """.format(url=url)
  display(HTML(html))
```
